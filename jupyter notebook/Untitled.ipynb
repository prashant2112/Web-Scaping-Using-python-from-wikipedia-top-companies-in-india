```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_India'
page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_tables_titles = [title.text.strip() for title in world_titles]
print(world_tables_titles)
```

```python
final_table_colume=world_tables_titles
print(final_table_colume)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns =  final_table_colume)
df
```

```python
column_data = table.find_all('tr')
```

```python
print(column_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    india_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = india_row_data



```

```python
df
```
